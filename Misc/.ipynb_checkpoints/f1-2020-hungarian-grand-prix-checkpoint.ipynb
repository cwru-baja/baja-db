# Introduction + Background
This is a notebook that generates a ridgeline plot of the laptimes for teams in the endurance race at Baja SAE Louisville 2021.

Data is available from the Baja SAE results website. I used a Chrome extension to download all the data and put it in an excel sheet, which is attached in this repo.

 Code based on F1 lap analysis code from reddit user hvcno.

```python
import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Helpers

```python
# Some regex helpers
driver_name = re.compile('([a-zA-Z])+\w ([A-Z])\w+')
lap_time = re.compile('([0-9])+:([0-9])+.([0-9])+')
lap_count = re.compile('([0-9])+')
```

```python
def convert_lap_time_to_seconds(x):
    tot_time = 3600*x.hour + 60*x.minute + x.second + 0.000001*x.microsecond
    #print(tot_time)
    return tot_time
```

# Execution

```python
excel_sheet1 = pd.read_excel('All Louisville Times Sheet.xlsx', "Virginia Tech")
excel_sheet2 = pd.read_excel('All Louisville Times Sheet.xlsx', 1)

teams = pd.ExcelFile('All Louisville Times Sheet.xlsx').sheet_names
# print(excel_file)
print(excel_sheet['Lap Time'])
print(teams)
```

```python
all_times = pd.DataFrame(columns=['Team Name', 'Lap Time'])

for team in teams:
    excel_sheet = pd.read_excel('All Louisville Times Sheet.xlsx', team)
    times = excel_sheet['Lap Time']

    if len(times) > 40:
        names_list = [team] * len(times)

        temp_frame = pd.DataFrame()
        temp_frame['Team Name'] = pd.Series(names_list)
        temp_frame['Lap Time'] = times

        all_times = all_times.append(temp_frame, ignore_index=True)

all_times = all_times.dropna()
print(all_times)
```

```python
all_times['laptime_seconds'] = all_times['Lap Time'].apply(convert_lap_time_to_seconds)
all_times = all_times[all_times['laptime_seconds'] < 300]  # Eliminate really long laptimes
```

# Plot
Adapted from an [official example](https://seaborn.pydata.org/examples/kde_ridgeplot.html)

```python
sns.set(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

g = sns.FacetGrid(all_times,
                  row='Team Name', # <-- "driver" column (was lazy and didn't rename)
                  hue='Team Name',
                  aspect=15,
                  height=1,
                  palette="viridis",
                  row_order=finishing_order,
                  hue_order=finishing_order)
# Map a KDE plot for each driver
g.map(sns.kdeplot, 'laptime_seconds', clip_on=False, shade=True, alpha=1, bw=0.5)
# Add a white outline
g.map(sns.kdeplot, 'laptime_seconds', clip_on=False, color='w', alpha=1, lw=3, bw=0.5)
# Add a solid horizontal line
g.map(plt.axhline, y=0, lw=2, clip_on=False)

# Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .1, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)


g.map(label, "Team Name")

# Set the subplots to overlap
g.fig.subplots_adjust(hspace=-0.5)

# Remove axes details that don't play well with overlap
g.set(yticks=[])
g.despine(bottom=True, left=True)

g.set_titles("")
g.set(xlim=(190,300))
g.set_axis_labels('Laptime (seconds)')
plt.suptitle('Laptime distributions for Baja SAE Louisville 2021', fontsize=20)
```
